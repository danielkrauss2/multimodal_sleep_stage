```python
def conf_matrix(confusion_matrix):
    x = confusion_matrix.sum(axis=1)[:, np.newaxis]
    confusion_matrix /=x
    confusion_matrix *= 100
    confusion_matrix = round(confusion_matrix,2)
    return confusion_matrix
```

```python
stage = "5stage"
mod = "acc"
```
